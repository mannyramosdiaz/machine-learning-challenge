```python
# Set the seed value for the notebook so the results are reproducible
from numpy.random import seed
seed(1)

import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense
import tensorflow
tensorflow.keras.__version__
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
data = pd.read_csv('../Resources/cumulative.csv')
data.set_index(['rowid'])
data.head()
```

```python
from sklearn.datasets import make_classification

X = data.drop("kepoi_name",  axis=1)
X = X.drop("kepler_name",  axis=1)
X = X.drop("koi_disposition",  axis=1)
X = X.drop("koi_pdisposition",  axis=1)
X = X.drop("koi_tce_delivname", axis=1)
X = X.drop("kepid", axis=1)
#X = X.dropna()
X = X
y = data["kepoi_name"]
y = y.values.reshape(-1, 1)
#y = y.dropna()
#y = y.values
print(X.shape, y.shape)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
X_train = X_train.values
X_test = X_test.values
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
# Create model and add layers
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=44))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=3, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=1
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:5])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:5])}")
```
